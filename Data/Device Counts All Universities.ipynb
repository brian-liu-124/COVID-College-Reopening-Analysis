```python
#load data
cbg_device_count = pd.read_csv('cbg_device_counts.csv')
cbg_device_count.drop('Unnamed: 0',axis = 1, inplace = True)
cbg_device_count['date'] = pd.to_datetime(cbg_device_count['date']).dt.date
college_list = pd.read_csv('college_list.csv')
centers = pd.read_csv('cbg_geos_centers.csv')
cbg_shapes = gpd.read_file('CBG_GEOS.csv')
college_details = pd.read_csv('College Details.csv')
reopening_status = pd.read_csv('ReopeningStatusFinal.csv')
```

```python
cbg_geo_map = pd.merge(cbg_device_count, centers[['CensusBlockGroup','lat','long']],left_on = 'origin_census_block_group', right_on = 'CensusBlockGroup', how = 'left')
cbg_geo_map = cbg_geo_map[['origin_census_block_group','lat','long']].drop_duplicates()
#For each university select all CBG within 2 mile radius
results = []
for i in range(0,len(college_list)):
    if i % 100 == 0:
        print(i)
    temp = college_list.iloc[i]
    lat = temp['Latitude']
    long = temp['Longitude']
    try:
        lat_ul = lat + 0.3
        lat_ll = lat - 0.3
        long_ul = long + 0.3
        long_ll = long - 0.3
        cond = (cbg_geo_map['lat'] < lat_ul) & (cbg_geo_map['lat'] > lat_ll) & (cbg_geo_map['long'] < long_ul) & (cbg_geo_map['long'] > long_ll)
        geo_list = cbg_geo_map[cond]
        geo_list['dist'] =  [haversine(x,(lat,long),unit=Unit.MILES) for x in list(zip(geo_list.lat, geo_list.long))]
        geo_list = geo_list[geo_list['dist'] <= 1]

        #device_count = sdm.groupby('date_range_start').device_count.sum().reset_index()
        
        #device_count['year'] =  device_count['date_range_start'].dt.year
        #device_count = device_count.groupby('year').device_count.mean()
        results.append( [temp['UnitID'],temp['Institution Name'],lat,long,geo_list['origin_census_block_group'].values] )
    except:
        results.append([temp['UnitID'],temp['Institution Name'],lat,long, None])
        
# 1.5 mile radius around school center for CBG
schools_cbg = pd.DataFrame(results,columns = ['unitid','inst_name','lat','long','cbg_in_radius'])
schools_cbg = schools_cbg[schools_cbg['unitid'].isin(college_details[college_details['Full-time undergraduate enrollment (DRVEF2018)']>10000]['UnitID'])]
        
```

### TO GeoJson

```python
geos = []
schools = []
results = []
status = []
for i,row in schools_cbg.iterrows():
    name = row['inst_name']
    id1 = row['unitid']
    try:
        status1 = reopening_status[reopening_status['UnitID'] == id1]['Reopening Status'].values[0]
    except:
        status1 = 'Unknown'
    if len(status1) == 0:
        status1 = 'Unknown'
    for j in row['cbg_in_radius']:
        poly = cbg_shapes[cbg_shapes['CensusBlockGroup'] == str(j).zfill(12)]['geometry'].values[0]
        geos.append(poly)
        schools.append(name)
```

```python
from geopandas import GeoSeries
from geopandas import GeoDataFrame
d = {'name':schools,'status':status,'geometry':geos}
geos1 = GeoDataFrame(d)
geos1.to_file("cbg_colleges.geojson", driver='GeoJSON')
```

```python
school_nbhd_counts = pd.DataFrame()
for school in schools_cbg['inst_name'].unique():
    cbg_int = schools_cbg[schools_cbg['inst_name']==school]['cbg_in_radius'].values[0]
    toplot = cbg_device_count[cbg_device_count['origin_census_block_group'].isin(cbg_int)].groupby('date').max_at_home.sum().reset_index()
    toplot['Name'] = str(school)
    toplot = pd.DataFrame(toplot)
    school_nbhd_counts=school_nbhd_counts.append(toplot)
```
